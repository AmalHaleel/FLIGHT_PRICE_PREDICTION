```python
#to fill NA we have to filter the rows and then replace NA using mode of the filtered ones
filtered=(data["Source"] == "Delhi") & (data["Destination"] == "Cochin")
route_mode=data.loc[filtered,"Route"].mode()[0]
stop_mode=data.loc[filtered,"Total_Stops"].mode()[0]
data.loc[filtered,"Route"]=data.loc[filtered,"Route"].fillna(route_mode)
data.loc[filtered,"Total_Stops"]=data.loc[filtered,"Total_Stops"].fillna(stop_mode)
```

```python
data['Arrival_Time']=data['Arrival_Time'].str.split().str[0]
```

```python
data.head()
```

```python
data['Arrival_Time']=pd.to_datetime(data['Arrival_Time'],format='%H:%M')
data['Arrival_hr']=pd.to_datetime(data['Arrival_Time']).dt.hour
data['Arrival_min']=pd.to_datetime(data['Arrival_Time']).dt.minute
data.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
data
```

```python
data['Dep_Time']=pd.to_datetime(data['Dep_Time'],format='%H:%M')
data['Dep_hour']=pd.to_datetime(data['Dep_Time']).dt.hour
data['Dep_min']=pd.to_datetime(data['Dep_Time']).dt.minute

def time_of_day(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

data["Dep_time_category"] = data["Dep_hour"].apply(time_of_day)



data.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.shape
```

```python
data['Date_of_Journey']=pd.to_datetime(data['Date_of_Journey'],format='%d/%m/%Y')
data['journey_day']=pd.to_datetime(data['Date_of_Journey']).dt.day
data['journey_month']=pd.to_datetime(data['Date_of_Journey']).dt.month
data['journey_year']=pd.to_datetime(data['Date_of_Journey']).dt.year
data["is_weekend"] = pd.to_datetime(data["Date_of_Journey"]).dt.weekday >= 5
data["is_weekend"] = data["is_weekend"].map({True: "Weekend", False: "Weekday"})
data.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
data.drop(['journey_year'],axis=1,inplace=True)
```

```python
def duration_to_minutes(duration):
    duration=duration.strip()
    hours=0
    mins=0

    if 'h' in duration:
        hours=int(duration.split('h')[0])
        duration=duration.split('h')[1]
    if 'm' in duration:
        mins=int(duration.split('m')[0].strip()) if duration.split('m')[0].strip()!='' else 0
    return hours*60+mins
data['Duration_mins']=data['Duration'].apply(duration_to_minutes)
    
```

```python
data.drop(['Duration'],axis=1,inplace=True)
```

```python
data["Total_Stops"] = data["Total_Stops"].replace({
    "non-stop": 0,
    "1 stop": 1,
    "2 stops": 2,
    "3 stops": 3,
    "4 stops": 4
})
data["Total_Stops"] = data["Total_Stops"].astype(int)
```

```python
data.head()
```

```python
# Convert both departure and arrival times into "minutes since midnight"
dep_total_minutes = data["Dep_hour"] * 60 + data["Dep_min"]
arr_total_minutes = data["Arrival_hr"] * 60 + data["Arrival_min"]

# If arrival minutes < departure minutes, it means the flight crossed midnight
data["is_overnight"] = (arr_total_minutes < dep_total_minutes).astype(int)
```

```python
Q1 = data["Price"].quantile(0.25)
Q3 = data["Price"].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = data[(data["Price"] < lower_limit) | (data["Price"] > upper_limit)]
print("Number of outliers:", outliers.shape[0])
```

```python
data["Price"] = np.log1p(data["Price"])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
x=data.drop(['Price'],axis=1)
y=data['Price']

categorical_col=x.select_dtypes(include=['object']).columns
numeric_col=x.select_dtypes(exclude=['object']).columns

categorical_transformer=OneHotEncoder(handle_unknown='ignore')
numeric_tranformer=StandardScaler()

preprocessor=ColumnTransformer(transformers=[('cat',categorical_transformer,categorical_col),('num',numeric_tranformer,numeric_col)])

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#liner regression

lin_pipeline=Pipeline(steps=[
    ("preprocessing",preprocessor),
    ('model',LinearRegression())
])

lin_pipeline.fit(x_train,y_train)

y_pred_lin=lin_pipeline.predict(x_test)

def evaluate_model(name, y_true, y_pred):
    print(f"{name} Performance:")
    print("R² Score :", round(r2_score(y_true, y_pred), 4))
    print("MAE      :", round(mean_absolute_error(y_true, y_pred), 2))
    print("RMSE     :", round(mean_squared_error(y_true, y_pred), 2))

evaluate_model("Linear Regression", y_test, y_pred_lin)
```

```python
#random forest regressor
```

```python
rf_pipeline=Pipeline(steps=[("preprocessing",preprocessor),
                            ("model",RandomForestRegressor(n_estimators=300,max_depth=None,verbose=1,n_jobs=-1,random_state=42))
                           ])
rf_pipeline.fit(x_train,y_train)

y_pred_rf=rf_pipeline.predict(x_test)

def evaluate_model(name, y_true, y_pred):
    print(f"{name} Performance:")
    print("R² Score :", round(r2_score(y_true, y_pred), 4))
    print("MAE      :", round(mean_absolute_error(y_true, y_pred), 2))
    print("RMSE     :", round(mean_squared_error(y_true, y_pred), 2))

evaluate_model("RandomForestRegressor", y_test, y_pred_rf)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

scores = cross_val_score(rf_pipeline, x, y, cv=5, scoring='r2')
print("CV Mean R²:", np.mean(scores))
print("CV Std Dev:", np.std(scores))
```

```python
from xgboost import XGBRegressor

xgb_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),  
    ('model', XGBRegressor(
        n_estimators=500,
        learning_rate=0.05,
        max_depth=7,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        n_jobs=-1
    ))
])

xgb_pipeline.fit(x_train,y_train)

y_pred_xgb=xgb_pipeline.predict(x_test)

def evaluate_model(name, y_true, y_pred):
    print(f"{name} Performance:")
    print("R² Score :", round(r2_score(y_true, y_pred), 4))
    print("MAE      :", round(mean_absolute_error(y_true, y_pred), 2))
    print("RMSE     :", round(mean_squared_error(y_true, y_pred), 2))

evaluate_model("XGBRegressor", y_test, y_pred_rf)
```

```python
from sklearn.linear_model import Ridge



ridge_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor), 
    ('model', Ridge(alpha=1.0, random_state=42))
])


ridge_pipeline.fit(x_train, y_train)


y_pred = ridge_pipeline.predict(x_test)


def evaluate_model(name, y_true, y_pred):
    print(f"{name} Performance:")
    print("R² Score :", round(r2_score(y_true, y_pred), 4))
    print("MAE      :", round(mean_absolute_error(y_true, y_pred), 2))
    print("RMSE     :", round(mean_squared_error(y_true, y_pred), 2))

evaluate_model('Ridge_Regressor', y_test, y_pred_rf)
```
